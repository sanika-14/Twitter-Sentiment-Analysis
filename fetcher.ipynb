```python
import tweepy
import pandas as pd
import time

# Twitter API credentials
bearer_token = 'AAAAAAAAAAAAAAAAAAAAAE1TxwEAAAAAGJsfxYZdLi9SGD88Roy85W6emZE%3D5XPEBfBjSPovWY1SE7V6LjSBlOIBcsr9b3f28rsagUW6pX7B3R'

# Authenticate to Twitter
client = tweepy.Client(bearer_token=bearer_token, wait_on_rate_limit=True)
search_query = "(world chess championship) (happy OR sad OR love OR hate OR amazing OR terrible OR great OR awful) lang:en"
no_of_tweets = 100

def fetch_tweets(client, query, max_results):
    tweets = []
    next_token = None
    while len(tweets) < max_results:
        try:
            response = client.search_recent_tweets(query=query, max_results=min(100, max_results - len(tweets)), tweet_fields=['created_at', 'public_metrics', 'source', 'text'], user_fields=['username'], next_token=next_token)
            if response.data:
                tweets.extend(response.data)
            if 'next_token' in response.meta:
                next_token = response.meta['next_token']
            else:
                break
        except tweepy.TooManyRequests:
            print("Rate limit exceeded. Sleeping for 15 minutes.")
            time.sleep(15 * 60)  # Sleep for 15 minutes
        except Exception as e:
            print('Status Failed On,', str(e))
            break
    return tweets

tweets_data = fetch_tweets(client, search_query, no_of_tweets)

if tweets_data:
    # Pulling some attributes from the tweet
    attributes_container = [[tweet.author_id, tweet.created_at, tweet.public_metrics['like_count'], tweet.source, tweet.text] for tweet in tweets_data]

    # Creation of column list to rename the columns in the dataframe
    columns = ["User ID", "Date Created", "Number of Likes", "Source of Tweet", "Tweet"]

    # Creation of Dataframe
    tweets_df = pd.DataFrame(attributes_container, columns=columns)
    print(tweets_df)
else:
    print("Failed to fetch tweets.")
```
